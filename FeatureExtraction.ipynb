```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r'/content/sample_data/anime.csv')
```

```python
df.head()
```

```python
df.loc[2]['Title']
```

```python
def extract_episode(text):

  check = False
  data = ""
  for i in text:
    if i == ')':
      check = False
      break
    if check == True:
      data = data + i
    if i == '(':
      check = True
  return data
```

```python
df['Episodes'] = df['Title'].apply(extract_episode)
```

```python
df['Episodes'] = df['Episodes'].str.replace(" eps","")
```

```python
df['Episodes'] = df['Episodes'].astype(int)
```

```python
def extract_time(txt):
  check = False
  data = ""
  for i in range(len(txt)):
    if txt[i] == ")":
      for j in range(i+1,i+20):
        data += txt[j]
      return data
```

```python
df['Total Time'] = df['Title'].apply(extract_time)
```
